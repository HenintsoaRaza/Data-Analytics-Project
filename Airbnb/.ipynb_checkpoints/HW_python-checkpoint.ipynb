```python
columns_listings = [ "id",'city', "neighbourhood_cleansed",
                    "latitude", "longitude",
                    "property_type", "room_type", "accommodates", "bedrooms",
                    "beds", "price", "minimum_nights",  "maximum_nights"]
```

```python
  ## arrange by id and date
    #calendar <- calendar %>% 
    #    arrange(listing_id, date)
calendar_malaga=calendar_malaga.sort_values(by=['listing_id','date'])
calendar_malaga
```

```python
 ## add day number (starting first day)
   # calendar <- calendar %>%
   #     group_by(listing_id) %>%
   #     mutate(day_nb = row_number()) %>%
   #     ungroup()
#calendar_malaga[day_nb]=
calendar_malaga['date'] = pd.to_datetime(calendar_malaga['date'])

calendar_malaga['day_nb']=calendar_malaga.groupby('listing_id')['date'].rank(method='first')
regarder=['listing_id','date','day_nb']
print(calendar_malaga['listing_id'].unique())
```

```python
calendar_malaga.loc[calendar_malaga['listing_id']==96033,regarder]
```

```python
calendar_malaga.loc[calendar_malaga['listing_id']==131725,regarder]
```

```python
#calendar <- calendar %>%
#      mutate(available = ifelse(available=="t", 1, 0))
calendar_malaga.loc[calendar_malaga['available']=='t','available']=1
calendar_malaga.loc[calendar_malaga['available']=='f','available']=0
calendar_malaga['available']
```

```python
#calendar <- calendar %>%
#        mutate(price = str_replace(price, "\\$", ""),
#               adjusted_price = str_replace(adjusted_price, "\\$", ""))
calendar_malaga['price']=calendar_malaga['price'].str.replace('\$','')
calendar_malaga['adjusted_price']=calendar_malaga['adjusted_price'].str.replace('\$','')
calendar_malaga[['price','adjusted_price']]
```

```python
# calendar <- calendar %>%
#        mutate(price = str_replace(price, ",", ""),
#               adjusted_price = str_replace(adjusted_price, ",", ""))
calendar_malaga['price']=calendar_malaga['price'].str.replace('\,','')
calendar_malaga['adjusted_price']=calendar_malaga['adjusted_price'].str.replace('\,','')
calendar_malaga[['price','adjusted_price']]
```

```python
#calendar <- calendar %>%
#        mutate(price = as.numeric(price),
#               adjusted_price = as.numeric(adjusted_price))
calendar_malaga['price']=pd.to_numeric(calendar_malaga['price'])
calendar_malaga['adjusted_price']=pd.to_numeric(calendar_malaga['adjusted_price'])
```

```python
## calculate estimated revenue for upcoming day
#    calendar <- calendar %>%
#        mutate(revenue = price*(1-available))
calendar_malaga['available']=pd.to_numeric(calendar_malaga['available'])
calendar_malaga['revenue']=calendar_malaga['price']*(1-calendar_malaga['available'])
calendar_malaga['revenue']
```
